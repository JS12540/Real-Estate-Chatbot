```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("image-to-text", model="jinhybr/OCR-Donut-CORD")
```

```python
# Function to use the pipeline on an image
def image_to_text(image_path):
    # Perform the image-to-text conversion
    result = pipe(image_path)
    return result[0]['generated_text']
```

```python
# Convert the image to text using the pipeline
text = image_to_text("/content/Tembusu grand 1 Bed + Study unit plan.png")
print("Extracted Text:", text)
```

```python
import re

def clean_extracted_text(text):
    # Remove tags like <s_nm>, <s_num>, etc.
    clean_text = re.sub(r'<[^>]+>', '', text)

    # Replace multiple spaces or newline characters with a single space
    clean_text = re.sub(r'\s+', ' ', clean_text)

    # Adding line breaks or formatting for readability
    formatted_text = clean_text.replace(":", ": ").replace("*", " ")
    formatted_text = re.sub(r'(#\d{2}-\d{2}to)', r'\n\1', formatted_text)  # Add line breaks before unit ranges
    formatted_text = re.sub(r'(SAM)', r'\1\n', formatted_text)  # Add line breaks after certain keywords
    formatted_text = re.sub(r'(TOTAL|DISCOUNT|PRICE|TYPE|AVAILABO|SHRATA|BALCOMY)', r'\n\1:', formatted_text)  # Separate sections
    formatted_text = re.sub(r'(,\s)', r'\n', formatted_text)

    return formatted_text.strip()
```

```python
formatted_text = clean_extracted_text(text)
```
